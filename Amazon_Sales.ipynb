###  Identifying the months having the maximum orders

```python
df1=df['Date'].value_counts().reset_index().sort_values(by='count',ascending=False).head()
sns.barplot(x='Date',y='count',palette='Set2',data=df1)
plt.title('Maximum orders by month')
```

#### It can be clearly seen that Feb 2022 is having the maximum orders followed by March.

###  Identifying the months having the minimum orders

```python
df1=df['Date'].value_counts().reset_index().sort_values(by='count',ascending=False).tail()
sns.barplot(x='Date',y='count',palette='Set2',data=df1)
plt.title('Minimum orders by month')
```

#### It can be clearly seen that March 2022 is having the minimum orders compared to other months 

###  Identify the top 2 expensive products

```python
top_2_expensive_products = df.groupby('Category')['Amount'].max().nlargest(2).index.tolist()
df_top_2_expensive_products = df[df['Category'].isin(top_2_expensive_products)]
plt.figure(figsize=(8, 4))
sns.barplot(x='Category', y='Amount', data=df_top_2_expensive_products,palette='Set2')
plt.title('Top 2 Expensive Products')
plt.show()
```

#### It clearly states that Set is more expensive than Western Dress

###  Identify the most shipped products from the above dataset

```python
product_counts = df['Category'].value_counts().reset_index()
product_counts.columns = ['Category', 'Date']
top_products = product_counts.head()
plt.figure(figsize=(8, 4))
sns.barplot(x='Category', y='Date', data=top_products, palette='Set2')
plt.xlabel('Number of Shipments')
plt.ylabel('Category')
plt.title(' Most Shipped Products')
plt.show()
```

#### It clearly states that set and kurta are most shipped products compared to Western dress ,Top and Ethinic dress

###  Identify top 5 state which makes the highest order


```python
state_order_counts = df.groupby('Ship-state').size().reset_index(name='Order Count')
state_order_counts_sorted = state_order_counts.sort_values(by='Order Count', ascending=False)
top_5_states = state_order_counts_sorted.head(5)
plt.figure(figsize=(6, 6))
plt.pie(top_5_states['Order Count'], labels=top_5_states['Ship-state'], autopct='%1.1f%%', startangle=140)
plt.title('Top 5 States with the Highest Orders')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

#### It clearly states that the state Maharashtra makes the highest order compared to other states

### Identify the most prefered ship service level

```python


service_level_counts = df['Ship-service-level'].value_counts().reset_index()
service_level_counts.columns = ['Ship-service-level', 'ShipmentCount']
service_level_counts = service_level_counts.sort_values(by='ShipmentCount', ascending=False)
plt.figure(figsize=(8, 4))
sns.barplot(x='ShipmentCount', y='Ship-service-level', data=service_level_counts, palette='Set2')
plt.title('Most Preferred Shipping Service Level')
plt.xlabel('No of shipments')
plt.ylabel('Shipping Service level')
plt.show()
```

#### It clearly states that expedited is the most preferred ship service level compared to standard

###  Identify the category which has highest sales during 2022

```python

category_sales = df.groupby('Category')['Amount'].sum().reset_index()
category_sales = category_sales.sort_values(by='Amount', ascending=False)
plt.figure(figsize=(8, 6))
sns.barplot(x='Category', y='Amount', data=category_sales, palette='Set2')
plt.title('Highest Sales Category in 2022')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.show()
```

#### It clearly states that category set has highest sales in 2022 compared to other categories

###  Identify the top 5 cities which makes highest order compared to others

```python
city_orders = df.groupby('Ship-city')['Order_id'].count().reset_index()

# Sort the cities by the number of orders in descending order
city_orders_sorted = city_orders.sort_values(by='Order_id', ascending=False)
top_5_cities = city_orders_sorted.head(5)
plt.figure(figsize=(9,6))
sns.barplot(x='Ship-city', y='Order_id', data=top_5_cities, palette=sns.color_palette("Set2"))
plt.title('Top 5 Cities with the Highest Number of Orders')
plt.xlabel('Ship-city')
plt.ylabel('Number of Orders')
plt.show()
top_5_cities


```

#### It clearly states that Bengaluru city makes the highest order compared to other cities

###  Identify the most unshipped products from the above dataset

```python
unshipped_df = df[df['Courier_status'] == 'Unshipped']
unshipped_count = unshipped_df.groupby('Category').size().reset_index(name='Count')
sns.barplot(x='Category', y='Count', data=unshipped_count,palette='Set2')
plt.title('Most Unshipped Categories')
plt.xlabel('Category')
plt.ylabel('Number of Unshipped Orders')
plt.xticks(rotation=116015)
plt.show()
```

#### It clearly states that set is most unshipped category compared to other categories

###  Find out the total quantity of items sold per categry

```python
category_quantities = df.groupby('Category')['Qty'].sum().reset_index()

# Plotting the bar chart
plt.figure(figsize=(9,6))
plt.bar(category_quantities['Category'], category_quantities['Qty'], color=plt.cm.Set2(range(len(category_quantities))))
plt.title('Total Quantity of Items Sold per Category')
plt.xlabel('Category')
plt.ylabel('Total Quantity')
plt.show()
```

#### It clearly states that the total quantity of items sold per category.This visualization helps in understanding the distribution of sales across different categories

###  Find all different categories and their counts

```python
category_counts = df['Category'].value_counts().reset_index()
category_counts.columns = ['Category', 'Count']
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Count', data=category_counts,palette = "Set2")
plt.title('Unique Product Categories in Orders')
plt.xlabel('Product Category')
plt.ylabel('Count')
plt.show()
```

#### It clearly states that set , Kurta ,Western dress and Top are the most common unique product categories in order

###  Identify the Top 5 states which have received B2B orders

```python
b2b_orders = df[df['B2b'] == True]

# Group by state and count orders
top_states = b2b_orders['Ship-state'].value_counts().nlargest(5)

# Plot
plt.figure(figsize=(6, 6))
plt.pie(top_states, labels=top_states.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 States with B2B Orders')
plt.show()
```

####  It clearly states that  State Maharashtra have received more B2B orders compared to other states.

### Identify the fulfillment types for orders placed on 4/30/2022

```python
df['Date'] = pd.to_datetime(df['Date'])
orders_04302022 = df[df['Date'] == '2022-04-30']
plt.figure(figsize=(6, 4))
sns.countplot(data=orders_04302022, x='Fulfilment',palette ="Set2")
plt.title('Fulfillment Types for Orders Placed on 4/30/2022')
plt.xlabel('Fulfillment Type')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-labels for better readability if needed
plt.show()
```

#### It clearly states that Amazon is more fulfillment type than Merchant for orders placed on 4/30/2022

### Are there any orders with a status other than 'Shipped' or 'Shipped - Delivered to Buyer

```python
filtered_df = df[~df['Status'].isin(['Shipped', 'Shipped - Delivered to Buyer'])]

# Visualize the counts of these statuses using Seaborn's pie plot
plt.figure(figsize=(8, 6))
sns.countplot(data=filtered_df, x='Status',palette="Set2")
plt.title('Counts of Orders with Statuses Other Than Shipped or Shipped - Delivered to Buyer')
plt.xlabel('Status')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

#### It clearly states that there are many orders with a status other than 'Shipped' or 'Shipped - Delivered to Buyer

### Identify the top 5 different styles from above dataset

```python
top_styles = df['Style'].value_counts().head(5)
plt.figure(figsize=(6, 6))
plt.title('Top 5 Different Styles')
sns.set_palette('pastel')  
top_styles.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.ylabel('')  # Remove the y-label
plt.show()
```

#### It clearly states that JNE3739,JNE3405,SET268,J0341,J0230 are the top 5 different styles

### Identify top 3 orders shipped with the 'Expedited' service level

```python
expedited_orders = df[df['Ship-service-level'] == 'Expedited']
order_quantities = expedited_orders.groupby('Category')['Qty'].sum()
sorted_orders = order_quantities.sort_values(ascending=False)
top_3_orders = sorted_orders.head(3)
plt.figure(figsize=(6, 6))
plt.title('Top 3 Orders Shipped with Expedited Service Level')
top_3_orders.plot(kind='pie', autopct='%1.1f%%', cmap='Set3')
plt.ylabel('')
plt.show()
```

#### It clearly states that Set, Kurta and Western Dress are the top 3 orders shipped with the 'Expedited' service level
